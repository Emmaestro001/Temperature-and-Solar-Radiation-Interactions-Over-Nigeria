```python
#lINEAR REGRESSION PLOTS

#IMPORT ALL TABLES FOR EACH ZONE
```

```python
NC = pd.read_excel("RAD AND TEMP.xlsx", 'NC1')
NC.head()
```

```python
NE = pd.read_excel("RAD AND TEMP.xlsx", 'NE1')
NE.head()
```

```python
NW = pd.read_excel("RAD AND TEMP.xlsx", 'NW1')
NW.head()
```

```python
SE = pd.read_excel("RAD AND TEMP.xlsx", 'SE1')
SE.head()
```

```python
SW = pd.read_excel("RAD AND TEMP.xlsx", 'SW1')
SW.head()
```

```python
SS = pd.read_excel("RAD AND TEMP.xlsx", 'SS1')
SS.head()
```

```python
#WE NOW ASSIGN EACH MEAN VALUE OF RADIATION (H) AND TEMPERATURE (T) FOR EACH ZONE TO A VARIABLE 

NW_x = NW.Mean_T
NW_y = NW.Mean_H

NE_x = NE.Mean_T
NE_y = NE.Mean_H

NC_x = NC.Mean_T
NC_y = NC.Mean_H

SW_x = SW.Mean_T
SW_y = SW.Mean_H

SE_x = SE.Mean_T
SE_y = SE.Mean_H

SS_x = SS.Mean_T
SS_y = SS.Mean_H
```

```python
#ALL THE MEAN VALUES FOR EACH ZONE WILL BE PLOTTED IN ONE FIGURE

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, figsize=[10,6.666])

#NORTHWEST plot
axs[0, 0].scatter(NW_x, NW_y, marker="+")
axs[0, 0].set_title("NorthWest")
axs[0, 0].text(30,16, '(a)', fontsize=20)

NW_z = np.polyfit(NW_x, NW_y, 1)
NW_p = np.poly1d(NW_z)
axs[0,0].plot(NW_x,NW_p(NW_x),"r--")

#NORTHEAST plot
axs[0, 1].scatter(NE_x, NE_y, marker="+")
axs[0, 1].set_title("NorthEast")
axs[0, 1].text(30,16, '(b)', fontsize=20)

NE_z = np.polyfit(NE_x, NE_y, 1)
NE_p = np.poly1d(NE_z)
axs[0,1].plot(NE_x,NE_p(NE_x),"r--")

#NORTHCENTRAL plot
axs[0, 2].scatter(NC_x, NC_y, marker="+")
axs[0, 2].set_title("NorthCentral")
axs[0, 2].text(30,16, '(c)', fontsize=20)

NC_z = np.polyfit(NC_x, NC_y, 1)
NC_p = np.poly1d(NC_z)
axs[0,2].plot(NC_x,NC_p(NC_x),"r--")

#SOUTHWEST plot
axs[1, 0].scatter(SW_x, SW_y, marker="+")
axs[1, 0].set_title("SouthWest")
axs[1, 0].text(30,24, '(d)', fontsize=20)

SW_z = np.polyfit(SW_x, SW_y, 1)
SW_p = np.poly1d(SW_z)
axs[1,0].plot(SW_x,SW_p(SW_x),"r--")

# SOUTHEAST plot
axs[1, 1].scatter(SE_x, SE_y, marker="+")
axs[1, 1].set_title("SouthEast")
axs[1, 1].text(30,24, '(e)', fontsize=20)

SE_z = np.polyfit(SE_x, SE_y, 1)
SE_p = np.poly1d(SE_z)
axs[1,1].plot(SE_x,SE_p(SE_x),"r--")

# SOUTHSOUTH plot
axs[1, 2].scatter(SS_x, SS_y, marker="+")
axs[1, 2].set_title("SouthSouth")
axs[1, 2].text(30,24, '(f)', fontsize=20)

SS_z = np.polyfit(SS_x, SS_y, 1)
SS_p = np.poly1d(SS_z)
axs[1,2].plot(SS_x,SS_p(SS_x),"r--")


for ax in axs.flat:
    ax.set(xlabel='Temperature($^oC$)', ylabel='Radiation (MJ/m$^2$/day)')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    
plt.show()
```
